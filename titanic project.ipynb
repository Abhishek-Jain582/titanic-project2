```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df
```

```python
#total number of people survived
def count():
    count=0
    for i in df['Survived']:
        if i==1:
            count+=1
    return count
x=count()
print(x)
            
            
```

```python
#total number of male and female in the titanic
def gender():
    count_male=0
    count_female=0
    for i in df['Sex']:
        if i=='male':
            count_male+=1
        else:
            count_female+=1
    return (count_male,count_female)
male,female=gender()
print("Number of Males=%d "%male)
print("Number of Females=%d "%female)
```
